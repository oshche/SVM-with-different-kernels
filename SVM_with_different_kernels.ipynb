Изучение SVM  с разными ядрами

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np
```

```python
moons_points, moons_labels = make_moons(n_samples=500, noise=0.2, random_state=42)
plt.scatter(moons_points[:, 0], moons_points[:, 1], c=moons_labels)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions
```

```python
# Посмотрим на SVM с распространнеными ядрами:
# poly - полиномиальная функци заданной степени;
# rbf - радиальная базисная функция;
# sigmoid - сигмоидальная функция;
for kernel_name in ['poly', 'rbf', 'sigmoid'] :
    if kernel_name == 'poly':
        for deg in [1,2,3,4,5] :
            svm = SVC(kernel=kernel_name,degree=deg)
            # degree параметр используется только poly ядром
            svm.fit(moons_points,moons_labels)
            acc = accuracy_score(moons_labels,svm.predict(moons_points))

            plt_name = 'SVM_%s_d%s, acc=%.4f' % (kernel_name,deg,acc)
            plt.title(plt_name)
            plot_decision_regions(X=moons_points,y=moons_labels,clf=svm)
            plt.show()
    else:
        svm = SVC(kernel=kernel_name)
        svm.fit(moons_points,moons_labels)
        acc = accuracy_score(moons_labels,svm.predict(moons_points))
        plt_name = 'SVM_%s, acc=%.4f' % (kernel_name,acc)
        plt.title(plt_name)
        plot_decision_regions(X=moons_points,y=moons_labels,clf=svm)
        plt.show()
```

```python
from sklearn.datasets import make_circles

circles_points, circles_labels = make_circles(n_samples=500, noise=0.06, random_state=42)

points = np.vstack((circles_points*2.5 + 0.5, moons_points))
labels = np.hstack((circles_labels, moons_labels + 2)) 
# + 2 для отдельных moon
plt.figure(figsize=(5, 5))
plt.scatter(points[:, 0], points[:, 1], c=labels)
plt.show()
```
